```python
import matplotlib 
matplotlib.__version__
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

```python
_ = plt.plot([0,1],[0,3])
```

```python
plt.plot([0,1],[0,3], label='line', color='green')
plt.ylabel('oś Y')
plt.xlabel('oś X')
plt.title('wykres')
plt.legend()
```

```python
plt.plot([0,1,2],[0,3,2], label='line', color='green')
plt.ylabel('oś Y')
plt.xlabel('oś X')
plt.title('wykres')
plt.legend()
```

```python
x = np.arange(-3,3,0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
plt.plot(x, x_cos)
```

```python
plt.subplot(121)
plt.plot([0,1,2],[3,2,5])
plt.subplot(122)
plt.plot([0,1,2],[6,1,8],'r--')
```

```python
fig = plt.figure(figsize=(20,6))
plt.subplot(131)
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[6,1,8],'r--')
plt.subplot(133)
plt.plot([0,1,2],[6,1,8],'ro')
```

```python
plt.bar(x=[0,1,2,3], height=[1.2,3.2,1.2,5.2])
```

```python
plt.barh(y=['a','b','c','d'], width=[1.2,3.2,1.2,5.2])
```

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)
plt.scatter(x=x1, y=x2)
```

```python
x3 = np.random.randn(300)*50
plt.scatter(x=x1, y=x2, s=x3)
```

```python
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5)
```
