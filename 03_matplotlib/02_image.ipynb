```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('office.jpg')
img
```

```python
img.shape
```

```python
plt.imshow(img)
plt.axis('Off')
```

```python
fig = plt.figure(figsize=(12,10))
plt.subplot(221)
plt.imshow(mpimg.imread('office.jpg'))
plt.subplot(222)
plt.imshow(mpimg.imread('business.jpg'))
plt.subplot(223)
plt.imshow(mpimg.imread('workplace.jpg'))
plt.subplot(224)
plt.imshow(mpimg.imread('children.jpg'))
```

```python
fig = plt.figure(figsize=(12,10))

for idx, img in enumerate(['office.jpg','business.jpg','workplace.jpg','children.jpg']):
  plt.subplot(221 + idx)
  plt.imshow(mpimg.imread(img))
```
